```python
import pandas as pd
```

# Data Cleaning
## training data

```python
train = pd.read_csv('./data/train_classes.csv',index_col = 0)
```

```python
train.columns = ['filename','not_hard_hat','hard_hat']
```

```python
train.sort_values(by = 'filename',ascending = True, inplace = True)
```

```python
train.reset_index(drop = True, inplace = True)
```

```python
train
```

```python
both = []

for x in range(len(train)):
    both.append(train['not_hard_hat'][x] + train['hard_hat'][x] - 1)

train['target'] = both
```

```python
train['target'] = train['target'] + train['not_hard_hat']
```

```python
train.head(8)
```

```python
train.to_csv('./data/train.csv')
```

## Moving files to the right folder

```python
files_0 = []
files_1 = []
files_2 = []

for x in range(len(train)):
    if train['target'][x] == 1:
        files_1.append(train['filename'][x])
    elif train['target'][x] == 2:
        files_2.append(train['filename'][x])
    else:
        files_0.append(train['filename'][x])
```
